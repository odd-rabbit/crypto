```python
s = '2020-08-02 02-AM'
def change_date(s):
    new_str = ''
    if (s[-2:] == 'AM' and s[11:] != '12-AM') or s[11:] == '12-PM' :
        new_str = s[:11] + str((int(s[11:13]))%24) + ':00'
    elif s[-2:] == 'PM' or s[11:] == '12-AM':
        new_str = s[:11] + str((int(s[11:13]) + 12)%24) + ':00'
    else:
        new_str = s
    new_str = pd.to_datetime(new_str)
    return new_str
change_date(s)
```
